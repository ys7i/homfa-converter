```python
bdict = spot.make_bdd_dict()
aut = spot.make_twa_graph(bdict)
p = buddy.bdd_ithvar(aut.register_ap("0"))
notp = buddy.bdd_ithvar(aut.register_ap("1"))
aut.new_states(3)
aut.set_init_state(0)
aut.new_edge(0, 1, p)
aut.new_edge(0, 2, notp)
aut
```

```python
formula = 'p & F(!p & Fp)'
tf = spot.formula(formula)
ff = spot.formula(f"!({formula})")
assert(tf.is_ltl_formula() and ff.is_ltl_formula)
tf_auto = spot.translate(tf, "BA")
tf_auto
```
